```python
import matplotlib.pyplot as plt
import numpy as np

def read_coefficients(filename):
    with open(filename, 'r') as f:
        # Split the line into individual numbers and convert each to float
        coefficients = [float(num) for num in f.read().split()]
    return coefficients

def quadratic_model(time, coefficients):
    a, b, c = coefficients
    temperature = a*(time**2) + b*time + c
    return temperature

def main():
    coefficients = read_coefficients('aishu2196.txt')
    time_values = np.linspace(0, 10, 50)
    temperature_values = quadratic_model(time_values, coefficients)
    plt.plot(time_values, temperature_values, label='Coefficients from file a=30 b=17 c=19')
    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather modeling with quadratic equation (coefficients from file)')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
  temperature = a*time**2 + b*time + c
  return temperature

def main():
  time_values = np.linspace(0, 10, 50)

  # Define three sets of coefficients
  coefficients_list = [(11, 15, 16), (2, 1, 13), (15, 2, 25)]

  for a, b, c in coefficients_list:
    temperature_hardcoded = quadratic_model(time_values, a, b, c)
    plt.plot(time_values, temperature_hardcoded, label=f'a={a}, b={b}, c={c}')

  plt.xlabel('Time')
  plt.ylabel('Temperature')
  plt.legend()
  plt.title('Weather Modelling with quadratic equation (Hard-coded coefficients)')
  plt.show()

if __name__ == '__main__':
  main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temperature_model(time, a, b, c):
    temperature = a * time**2 + b * time + c
    return temperature

file_paths = ['aishu2196.txt', '2504ai.txt', '0404ai.txt']

time_values = np.linspace(0, 50, 51)

for file_path in file_paths:
    try:
        with open(file_path, 'r') as file:
            values = file.readline().strip().split()
            if len(values) == 3:
                a = float(values[0])
                b = float(values[1])
                c = float(values[2])
                label = f'a={a}, b={b}, c={c}'
            else:
                raise ValueError("File should contain exactly 3 values.")

        temperatures = quadratic_temperature_model(time_values, a, b, c)
        plt.plot(time_values, temperatures, label=label)

    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
    except ValueError as e:
        print(f"Error: {e}")

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Weather Modelling with quadratic equation')
plt.legend()
plt.show()
```
